```python
olympics_df = pd.read_csv("olympics.csv", header=1)
olympics_df.head()
```

```python
new_names =  {'Unnamed: 0': 'Country',
               '? Summer': 'Summer Olympics',
               '01 !': 'Gold',
               '02 !': 'Silver',
               '03 !': 'Bronze',
               '? Winter': 'Winter Olympics',
               '01 !.1': 'Gold.1',
               '02 !.1': 'Silver.1',
               '03 !.1': 'Bronze.1',
               '? Games': '# Games',
               '01 !.2': 'Gold.2',
               '02 !.2': 'Silver.2',
               '03 !.2': 'Bronze.2'}
```

```python
olympics_df.rename(columns=new_names, inplace=True)
```

```python
olympics_df.head()
```

```python
import matplotlib.pyplot as plt 
x = olympics_df['Country'].loc[0:10] 
y = olympics_df['Combined total'].loc[0:10]

plt.pie( y, labels=x, autopct='%1.1f%%', ) 
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x = olympics_df['Country'].loc[0:10] 
y = olympics_df['Combined total'].loc[0:10]

x_coords = np.arange(len(x)) 
plt.bar(x_coords, y) 
plt.show()
```
